```python
# -*- coding: utf-8 -*-
"""
Created on Wed Nov  1 16:36:49 2017

@author: mj
"""
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img

a = img.imread("1.jpg")

def plot(plotnumbercount, image, title):
    subplot = fig.add_subplot(231 + plotnumbercount)
    subplot.imshow(image, cmap=plt.cm.gray)
    subplot.axis('off')
    subplot.set_title(title)
    
filter_mat  =  [[-1,-1,-1],[-1,8,-1],[-1,-1,-1]]

result = np.zeros((a.shape[0]-2, a.shape[1]-2))

for i in range(a.shape[0]-2):
    for j in range(a.shape[1]-2):
        result[i][j] = np.multiply(a[i:i+3,j:j+3], filter_mat).sum()
        
for i in range(result.shape[0]):
    for j in range(result.shape[1]):
        result[i][j] = max(result[i][j], 0)

fig = plt.figure(figsize=(15, 10))
plot(0, a, 'Original Image')
plot(1, result, 'Edge Filter')
plt.show()
```
